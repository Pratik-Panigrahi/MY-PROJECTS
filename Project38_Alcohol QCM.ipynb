```python
#chcking outliers
from scipy.stats import zscore
import numpy as np
z=np.abs(zscore(data))
z
```

```python
threshold=3
print(np.where(z>3))
df=data[(z<3).all(axis=1)]
df
```

```python
#Implementing k-means
kmeans=KMeans(n_clusters=5)
```

```python
kmeans.fit(df)
```

```python
pred=kmeans.predict(df)
```

```python
pred
```

```python
clusters=kmeans.cluster_centers_
print(clusters)
```

```python
pd.Series(pred).value_counts()
```

```python
kmeans.inertia_
```

```python
kmeans.score(df)
```

```python
SSE=[]
```

```python
for cluster in range(1,10):
    kmeans=KMeans(n_clusters=cluster,n_jobs=-1)
    kmeans.fit(df)
    SSE.append(kmeans.inertia_)
```

```python
a={'Cluster':range(1,10),'SSE':SSE}
frame=pd.DataFrame.from_dict(a, orient='index')
df1=frame.transpose()
```

```python
plt.figure(figsize=(12,6))
plt.plot(df1['Cluster'],df1['SSE'],marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('inertia')
plt.show()
```
